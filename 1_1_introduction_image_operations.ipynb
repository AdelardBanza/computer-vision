### Reading an Image

OpenCV allows reading different types of images (JPG, PNG, etc). You can load grayscale images colour images or you can also load images with Alpha channel. It uses the imread function which has the following code syntax arguments:

Reading function code Syntax arguments:

cv2.imread(    filename[, flags]    )

It has 2 arguments:

- Path of the image file: This can be an absolute or relative path. 
- Flags (optional): These flags are used to read an image in a particular format. The default is 1. 

Flag types:

- cv2.IMREAD_GRAYSCALE or 0: Loads image in grayscale mode
- cv2.IMREAD_COLOR or 1: Loads a color image. Any transparency of image will be neglected. It is the default flag.
- cv2.IMREAD_UNCHANGED or -1: Loads image as such including alpha channel.

```python
image_path = DATA_PATH + "/images/number_zero.jpg"

# Read image in Grayscale format, cv2.IMREAD_GRAYSCALE or 0:
sample_image = cv2.imread(image_path,0)
print(sample_image)
```

### Intensity 

The values printed above are the intensity values of each pixel.

0 means black pixel and as the value increases, it moves towards white. A value of 255 is a white pixel.

### Additional Display Utility Functions 
There are 4 more functions that are present in OpenCV which are commonly used with cv2.imshow function.

### 1. cv2.namedWindow

This function is used to create a display window with a specific name. This name is provided as the first argument of this function. The second argument is a flag which determines whether to resize the window to match the image size. The default autose. (cv2.WINDOW_NORMAL) and (cv2.WINDOW_AUTOSIZE - Default flag).

Function Syntax

image    =    cv2.namedWindow(    window[, flags]    )

Parameters

window - Name of the window in the window caption that may be used as an identifier.
flags - Flags of the window. The supported flags are: (cv::WindowFlags)

### 2. cv2.waitKey

This is a keyboard binding function. Its only argument is time in milliseconds. The function waits for specified milliseconds for any keyboard event. If you press any key in that time, the program continues. If 0 is passed, it waits indefinitely for a key stroke. It can also be set to detect specific key strokes which can be quite useful in video processing applications.

Function Syntax
Docs

waitkey    =    cv2.waitKey(    [, delay]    )
Parameters

delay - Delay in milliseconds. 0 is the special value that means "forever".

### 3. cv2.destroyWindow

This function is used to destroy or close a particular display window. The name of the window is provided as an argument to this function.

Function Syntax

destroywindow    =    cv2.destroyWindow(    window    )

Parameters

window - Name of the window to be destroyed

### 4. cv2.destroyAllWindows

This function is used to destroy all display windows. This function does not take any arguments.

Function Syntax

allwindows    =    cv2.destroyAllWindows(        )

### Working with color images¶
We know that an image is represented as a 2D Matrix. Here, we will see how to load color images and observe a few important points related to color images.

```python
# Read image
sample_color_image = DATA_PATH + "/images/face.jpg"

img = cv2.imread(sample_color_image)
print("Color image dimension ={}".format(img.shape))
```

The image in has 3 dimensions.
The third dimension indicates the number of channels in an image. This number is 3 ( namely R,G,B ). In some cases, there may be an additional channel (called alpha channel) which contains transparency information of the pixels

### Image Channels

The color image consists of multiple channels. Each channel itself is a grayscale image. The combination of intensity values of the three channels gives the color that is displayed on the screen. There are many color spaces used in practice. The most popular color space is the RGB color space.

In OpenCV, the order of channels R, G and B is reverse. i.e. In the image matrix, the Blue channel is indexed first, followed by the Green Channel and finally the Red Channel.

### Display the image

```python
# Display image
plt.imshow(img)

plt.show()
```

There is something weird about the image here because, OpenCV uses BGR format by default whereas Matplotlib assumes the image to be in RGB format. We can fix this either by converting the image to RGB colorspace using cv2.cvtColor(img, cv2.COLOR_BGR2RGB) or reverse the order of channels by swapping the 1st and 3rd channel - plt.imshow(img[:,:,::-1])

```python
# Convert BGR to RGB colorspace
imageRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(imageRGB)

plt.show()
```

```python
# Or we can also use the following by reversing the channels
plt.imshow(img[:,:,::-1])
```

Accessing the 3 channels and display the gray scale images in each channel

```python
# Show the channels
plt.figure(figsize=[20,5])

plt.subplot(131);plt.imshow(img[:,:,0]);plt.title("Blue Channel");
plt.subplot(132);plt.imshow(img[:,:,1]);plt.title("Green Channel");
plt.subplot(133);plt.imshow(img[:,:,2]);plt.title("Red Channel");
```

We can also work with the individual channels using the split and merge functions. 

- cv2.split()
- cv2.merge()

```python
# Splitting the image into the B, G, R formats
blue, green, red = cv2.split(img)

# Display the channels
plt.figure(figsize=[20, 5])

plt.subplot(141)
plt.imshow(blue)
plt.title("Blue Channel")

plt.subplot(142)
plt.imshow(green)
plt.title("Green channel")

plt.subplot(143)
plt.imshow(red)
plt.title("Red channel")

# Merge the channels into a BGR image
merged_image = cv2.merge((blue, green, red))

# Display the merged image
plt.subplot(144)
plt.imshow(merged_image[:, :, ::-1])
plt.title("Merged image")
```

### Modify Pixels

- location [0,0] to Yellow ( a mixture of Red and Green )
- location [1,1] to Cyan ( a mixture of Blue and Green )
- location [2,2] to Magenta ( a mixture of Blue and Red )

```python
plt.figure(figsize=[20, 20])

sample_image[0, 0] = (0, 255, 255)
plt.subplot(131)
plt.imshow(sample_image[:, :, ::-1])

sample_image[1, 1] = (255, 255, 0)
plt.subplot(132)
plt.imshow(sample_image[:, :, ::-1])

sample_image[2, 2] = (255, 0, 255)
plt.subplot(133)
plt.imshow(sample_image[:, :, ::-1])

plt.show()
```

### Images with Alpha Channel
There are 4 channels in color images, i.e. 3 color channels and 1 alpha channel which indicates the transparency value ( 0 - 255).

```python
# Read the image
panther_image = DATA_PATH + "/images/panther.png"

# -1 in cv2.imread() function tells the program to read the image as is
channel_image = cv2.imread(panther_image, -1)
print("Image shape ={}".format(channel_image.shape))

# combined channels to form BGR image
image_bgr_channel = channel_image[:, :, 0:3]

# alpha channel of the original image
image_mask = channel_image[:, :, 3]
plt.figure(figsize=[15, 15])
plt.subplot(121)
plt.imshow(image_bgr_channel[:, :, ::-1])
plt.title('Image color channel')

plt.subplot(122)
plt.imshow(image_mask, cmap='gray')
plt.title('Alpha channel')

plt.show()
```

The alpha channel is basic segmentation of the image. Which can be useful for creating overlays, i.e. augmentated reality applications.
