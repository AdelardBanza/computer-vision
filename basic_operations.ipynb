# Basic operations

Some basic operations on images which are used for any computer vision project. 

- Copying and creating new images
- Cropping an image section
- Resizing an Image
- Creating binary masks for images

```python
# importing libraries
import cv2
import numpy as np
from dataPath import DATA_PATH
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.rcParams['figure.figsize'] = (6.0, 6.0)
mpl.rcParams['image.cmap'] = 'gray'
```

### Creating new or copies of images

```python
# read image
sample_image = cv2.imread(DATA_PATH + "images/human_face.jpg")
```

```python
# displaying the image
plt.imshow(sample_image[:, :, ::-1])

plt.show()
```

### creating a copy of an image

```python
sample_image_copy = sample_image.copy()
```

```python
# displaying the copied image
plt.imshow(sample_image_copy[:, :, ::-1])
plt.show()
```

### Creating an empty matrix

```python
# creating an empty matrix with size (100, 200)
empty_matrix_image = np.zeros((100, 200, 3), dtype='uint8')
plt.imshow(empty_matrix_image)
plt.show() # optional
```

```python
# changing the empty image with white pixels
white_empty_matrix_image = 255 * np.ones((100, 200, 3), dtype='uint8')
plt.imshow(white_empty_matrix_image)
plt.show() # optional
```

### Creating an empty matrix of the same size as an existing image
Creating a matrix with the same size and data type as the existing image and fill it with 100 using the np.ones_like function

```python
empty_sample_image = 100 * np.ones_like(sample_image)
plt.imshow(empty_sample_image)
plt.show()
```

### Cropping an image section

```python
plt.imshow(sample_image[:, :, ::-1])
plt.show()
```

Let's crop the face. From the image we can see that the face is located somewhere in the region:

- between row   # 0 and row #300
- between colum # 150 and column # 400

So, for cropping out this area, wee need to select region from the image matrix and assign it to another variable. 

```python
cropped_image = sample_image[0:300, 150:400]
```

```python
plt.imshow(cropped_image[:, :, ::-1])
plt.show()
```

### Copying a cropped image

```python
copy_image = sample_image[1:280, 160:280]

# find height and width of the region of interest
region_height, region_width = copy_image.shape[:2]

# copy to left of face
sample_image_copy[40:40 + region_height, 10:10 + region_width] = copy_image

# copy to the right of face
sample_image_copy[40:40 + region_height, 350:350 + region_width] = copy_image

# display the output
plt.figure(figsize=[15, 15])
plt.subplot(121)
plt.imshow(sample_image[..., ::-1])
plt.title("Original image")

plt.subplot(122)
plt.imshow(sample_image_copy[..., ::-1])
plt.title("New image")
plt.show()
```
