### Resize an image

We are using cv2.resize function for resize an image.

Function Syntax 

dst    =    cv2.resize(    src, dsize[, dst[, fx[, fy[, interpolation]]]]    )

Parameters

src - input image
dst - output resized image
dsize - output image size
fx - scale factor along the horizontal axis;
fy - scale factor along the vertical axis; Either dsize or both fx and fy must be non-zero.
interpolation - interpolation method ( Bilinear / Bicubic etc ).

Check interpolation flags here 

https://docs.opencv.org/4.1.0/da/d54/group__imgproc__transform.html#ga5bb5a1fea74ea38e1a5445ca803ff121

There are two ways of using the resize function.

Specify width and height of output image explicitly

Use the dsize argument to specify the output size in the form (width,height).

Specify the scaling factors for resizing ( for both width and height )

Use fx and fy arguments to specify the scaling factor for horizontal and vertical axis respectively. It should be a float type.

The output size is calculated as: Size(round(fx*src.cols), round(fy*src.rows))

### Method1 - Specify width and height

We need to manually keep track of the aspect ratio of the output image as this method will not preserve the aspect ratio between input and output image.

Let's see how to use the functions in OpenCV for resizing.

```python
# rows and columns setup
resizeDownWidth = 300
resizeDownHeight = 200
resizeDown = cv2.resize(sample_image, (resizeDownWidth, resizeDownHeight), interpolation=cv2.INTER_LINEAR)

# Changing the aspect ratio
resizeUpWidth = 600
resizeUpHeight = 900
resizeUp = cv2.resize(sample_image, (resizeUpWidth, resizeUpHeight), interpolation=cv2.INTER_LINEAR)

plt.figure(figsize=[15, 15])
plt.subplot(131)
plt.imshow(sample_image[:, :, ::-1])
plt.title("Original sample image")

plt.subplot(132)
plt.imshow(resizeUp[:, :, ::-1])
plt.title("Scaled Up sample image")

plt.subplot(133)
plt.imshow(resizeDown[:, :, ::-1])
plt.title("Scaled Down sample image")

plt.show()
```

### Method2 - Specify scaling factor
This method preserve the aspect ratio of the image. Let's say we have an image of size 400x600 and we want the width to be 360. Then, instead of specifying the height, you can find the scaling factor (i.e 360/600 = 0.6) and use it to resize the image.

```python
# Scaling Up the image 1.5 times by specifying both scaling factors
scaleUpX = 1.5
scaleUpY = 1.5

# Scaling Down the image 0.6 times specifying a single scale factor
scaleDown = 0.6

scaledDown = cv2.resize(sample_image, None, fx= scaleDown, fy= scaleDown, interpolation= cv2.INTER_LINEAR)

scaledUp = cv2.resize(sample_image, None, fx= scaleUpX, fy= scaleUpY, interpolation= cv2.INTER_LINEAR) 
```

```python
# Displaying the image
plt.figure(figsize=[15, 15])
plt.subplot(121)
plt.imshow(scaledDown[..., ::-1])
plt.title("Scaled Down Sample Image, size = {}".format(scaledDown.shape[:2]))

plt.subplot(122)
plt.imshow(scaledUp[..., ::-1])
plt.title("Scaled Up Sample Image, size = {}".format(scaledUp.shape[:2]))

plt.show()
```

### Creating an Image Mask 
Masking is a very important step in many image processing and computer vision algorithms. The goal is to segment out the region of interest and apply the algorithm to a specific part of the image. We can perform all operations on this segmented part of the image and put it back on the original image.

### Create a mask using coordinates

```python
# Create an empty image of same size as the original sample image
sample_mask1 = np.zeros_like(sample_image)
plt.imshow(sample_mask1)
plt.show()
```

```python
sample_mask1[0:300, 150:400] = 255

plt.figure(figsize=[15, 15])
plt.subplot(121)
plt.imshow(sample_image[:, :, ::-1])
plt.title("Original Sample Image")

plt.subplot(122)
plt.imshow(sample_mask1[:, :, ::-1])
plt.title("Sample Image Mask")
plt.show()

print(sample_mask1.dtype)
```

This is the mask for the face region of the input image

### Create a mask using pixel intensity or color
Another way of creating masks is by using some logic. One example is using a simple color information. For instance, we can try to find out the pixels which are approximately red in color.

If we want to focus on red pixels, the simplest logic that does the trick is:

* The red channel should have high intensity ( keep the range of pixel values from 100 to 255 )
* The other 2 channels should have low intensity ( keep the range of pixel values in Blue and Green channels between 0 to 100)

There is a nice OpenCV function which can do exactly this. We will use the opencv function inRange

Function Syntax

It finds the pixels which lie in between the specified range. It produces a binary output image in which the white pixels corresspond to those pixels in the input image which fall in the specified range. The pixel values which fall outside the specified range are black (0)

dst    =    cv2.inRange(    src, lowerb, upperb[, dst]    )

Parameters

src - first input array.
lowerb - inclusive lower boundary array or a scalar.
upperb - inclusive upper boundary array or a scalar.
dst - output array of the same size as src and CV_8U type.

It produces a binary image ( pixels are either black or white ).

```python
sample_mask2 = cv2.inRange(sample_image, (0, 0, 150), (100, 100, 255))

plt.figure(figsize=[15, 15])
plt.subplot(121)
plt.imshow(sample_image[..., ::-1])
plt.title("Original Sample Image")

plt.subplot(122)
plt.imshow(sample_mask2)
plt.title("Sample Masked Image")
plt.show()
```

As we can see the white portion in the output mask corressponds to the dark pixels in the original image.
